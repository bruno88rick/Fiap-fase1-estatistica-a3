```python
#EXEMPLO 1 - DISTRIBUIÇÃO BEM COMPORTADA

# VERIFICAR A MEDIA / MEDIANA / MODA

#BIBLIOTECAS

import pandas as pd
```

```python
notas = pd.Series([6, 7, 7, 8, 8, 8, 9, 9, 10])
```

```python
print("Media: ", notas.mean())
print("Mediana: ", notas.median())
print("Moda: ", notas.mode().values)
```

```python
# EXEMPLO 2 - DISTRIBUIÇÃO COM OUTLIERS - DISTORCENDO A MÉDIA

salarios = pd.Series([1000, 1200, 1500, 2000, 2500, 3000, 3500, 4000, 50000])
```

```python
print('Média: ', salarios.mean())
print('Mediana: ', salarios.median())
```

```python
# Moda com Texto

produtos = pd.Series([
    'Coca-cola', 'Coca-cola', 'Pepsi', 'Fanta', 'Fanta', 'Sprite', 'Sprite', 'Sprite', "Pepsi"
])
```

```python
print("Moda: ", produtos.mode().values )
```

```python
# EXEMPLO 3

df = pd.DataFrame({"Notas": notas, "Salarios": salarios.reindex(range(len(notas)))})
```
