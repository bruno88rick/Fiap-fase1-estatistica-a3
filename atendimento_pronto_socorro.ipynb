```python
import pandas as pd
import numpy as np
dados = pd.Series([10,12,15,20,22,22,23,25,25,26,27,28,30,32,33,35,36,40,45,50,60,75,90,120,180])
# estatísticas
print("n =", len(dados))
print("soma =", dados.sum())
print("média =", dados.mean())
print("mediana =", dados.median())
# modo (pode ter múltiplos)
m = dados.mode()
print("moda(s) =", list(m))
print("Q1, Q3 =", dados.quantile([0.25, 0.75]).values)
print("IQR =", dados.quantile(0.75) - dados.quantile(0.25))
print("P90 =", dados.quantile(0.90))
print("% > 60 min =", (dados > 60).mean() * 100, "%")
# gráficos rápidos (requer matplotlib)
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
dados.hist(bins=10)
plt.title("Histograma - tempos de espera")
plt.xlabel("minutos")
plt.subplot(1,2,2)
dados.plot.box(vert=False)
plt.title("Boxplot - tempos de espera")
plt.show()
```
