```python
## bibliotecas

import pandas as pd
import numpy as np
```

```python
# Exemplo 1 - Amplitude
loja_a = pd.Series([100, 150, 200, 250, 300])
loja_b = pd.Series([50, 100, 150, 200, 250, 300, 350, 400, 450, 500])

def amplitude(dados):
    return dados.max() - dados.min()

print("Amplitude loja A:", amplitude(loja_a))
print("Amplitude loja B:", amplitude(loja_b))
```

```python
# Exemplo 2 - Variância

print("Variância loja A:", loja_a.var())
print("Variância loja B:", loja_b.var())
```

```python
# Exemplo 3 - Desvio Padrão
print("Desvio Padrão loja A:", loja_a.std())
print("Desvio Padrão loja B:", loja_b.std())
```

```python
# Exemplo 4 - Coefficientes de Variação - CV
def coeficiente_variacao(dados):
    media = dados.mean()
    desvio_padrao = dados.std()
    return desvio_padrao / media    

print("Coeficiente de Variação loja A:", coeficiente_variacao(loja_a))
print("Coeficiente de Variação loja B:", coeficiente_variacao(loja_b)) 
```

```python
print("Coeficiente de Variação loja A:", coeficiente_variacao(loja_a) * 100, "%")
print("Coeficiente de Variação loja B:", coeficiente_variacao(loja_b) * 100, "%") 
```

```python
# Dados
dados = [4, 6, 8, 5, 7]
n = len(dados)

# 1. Calcular média
media = sum(dados) / n
print("Média:", media)

# 2. Diferenças ao quadrado
quadrados = [(x - media) ** 2 for x in dados]
print("Diferenças ao quadrado:", quadrados)

# 3. Variância amostral
variancia = sum(quadrados) / (n - 1)
print("Variância:", variancia)
```

```python
# Variancia populacional

import numpy as np

dados = np.array([4, 6, 8, 5, 7])
variancia_pop = np.var(dados)      # divisor = N
print(variancia_pop)
```

```python
# variancia amostral
variancia_amostral = np.var(dados, ddof=1)   # divisor = n - 1
print(variancia_amostral)
```

```python
# variancia populacional
import numpy as np
dados = np.array([4, 6, 8, 5, 7])
variancia_pop = np.var(dados)      # divisor = N
print(variancia_pop)
```

```python
# variancia amostral
variancia_amostral = np.var(dados, ddof=1)    # divisor = n - 1
print(variancia_amostral)
```

```python
# desvio padrão

# Dados
dados = [4, 6, 8, 5, 7]
n = len(dados)
print("n:", n)

# 1. Média
media = sum(dados) / n
print("Média:", media)

# 2. Diferenças ao quadrado
quadrados = [(x - media) ** 2 for x in dados]
print("Diferenças ao quadrado:", quadrados)

# 3. Variância amostral
variancia = sum(quadrados) / (n - 1)
print("Variância:", variancia)

# 4. Desvio-padrão
desvio_padrao = variancia ** 0.5
print("Desvio-padrão:", desvio_padrao)
```

```python
# Este método mostra passo a passo como o desvio-padrão é realmente obtido.
# Dados
dados = [4, 6, 8, 5, 7]
n = len(dados)

# 1. Média
media = sum(dados) / n

# 2. Diferenças ao quadrado
quadrados = [(x - media) ** 2 for x in dados]

# 3. Variância amostral
variancia = sum(quadrados) / (n - 1)

# 4. Desvio-padrão
desvio_padrao = variancia ** 0.5
print("Desvio-padrão:", desvio_padrao)
```
