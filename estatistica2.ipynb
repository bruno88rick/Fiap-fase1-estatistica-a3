```python
# Bibliotecas utilizadas
import math
```

```python
# DEMONSTRAÇÃO 1 > AMOSTRA PARA PROPORÇÃO BINARIA

def amostra_proporcao(z, p, e):
    return math.ceil((z**2 * p * (1-p)) / e**2)
```

```python
z = 1.95 # 95% de confiança
p = 0.5 # proporção esperada (pior caso)
e = 0.05 # margem de erro de 5%

amostra_proporcao(z, p, e)
```

```python
# DEMOSNNNTRAÇÃO 2 > AMOSTRA POR MEDIA
def amostra_media(z, desvio_padrao, erro):
    return math.ceil((z * desvio_padrao / erro) ** 2)
```

```python
z= 1.96 # 95% de confiança
desvio = 12 # ex: tempo de atendimento de 12
erro = 2 # margem de erro de 2 minutos

amostra_media(z, desvio, erro)
```

```python
# demosntração 3 > Correção para população finita
def correcao_populacao_finita(n, N):
    return math.ceil((N * n) / (N + n - 1))
```

```python
# caso: ˜Tenho 5.000 clientres, quantas eu preciso amostra"
N = 5000 # população
n_bruto = 385
correcao_populacao_finita(n_bruto, N)
```
